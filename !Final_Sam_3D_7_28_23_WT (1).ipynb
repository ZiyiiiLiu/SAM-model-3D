## Define equations and parameters


```python
#DL=0.2  KCW2=0.08 KWC=0.25 KCL1=3.8   0.29  KCW3=0.2  KWW=0.25  gc=0.6 KWL=0.24   L_max=10.6;
#change DL and Lmax
```

```python
model_str = '''
    // Reactions:
    J00: -> W0;    kw*w0-bW*W0
    J01: -> c0;    -gc*degc*c0
    J02: -> C0;    kc*c0-1*C0
    J03: -> L0;    -bL*L0
    J04: -> w0;    -1*w0
    J05: -> H0;    -gH*H0

    // Variable Init:
    KCW1=1000; nCW1=5; 
    KCW2=0.11; nCW2=8; 
    KCL1=3.5; nCL1=6; gc=0.8;
    KCH1=0.3; nCH1=5; 
    kc=1; degc=0.8;
    KCW3=0.25; nCW3=3;

    k0WL=1; k0WC=0.7;  
    k0WW=1.5; KWW=0.3; nWW=2.5;
    KWL=0.23; nWL=5;
    KWC=0.265; nWC=5;
    kw=3.9*0.4; 
    bW=1; 
    aw0=0.31;

    KHW1=0.2; nHW1=4; gH=0.4;kH=0.5;KWH1=1;

    L_max=10.6; 
    bL=1.9;


    DW=6;DC=8;DL=0.2;DH=0.00;

'''


sam.add_basal_str(model_str)

#print(sam.model_str)
sam.expand_model()
#print(sam.model_str)
sam.set_init({'init_W':{'w':1.2, 'W':1.2}})
#sam.set_init({'init_W':{'w':0.3, 'W':0.2}})
#exit()

sam.add_syn2model('L', 'LR', 'L_max')
#sam.add_syn2model('c', 'all', 'gc/(1+(Lxxx/KCL1)^nCL1)*(Wxxx/KCW2)^nCW2/(1+(Wxxx/KCW2)^nCW2)/(3+((Wxxx-0.31)/KCW3)^nCW3)/(1+(Hxxx/KCH1)^nCH1)')
#sam.add_syn2model('c', 'all', 'gc/1*(Wxxx/KCW2)^nCW2/(1+(Wxxx/KCW2)^nCW2)/(3+((Wxxx-0.31)/KCW3)^nCW3)/(1+(Hxxx/KCH1)^nCH1)')
sam.add_syn2model('c', 'all', 'gc/(1+(Lxxx/KCL1)^nCL1)*(Wxxx/KCW2)^nCW2/(1+(Wxxx/KCW2)^nCW2)/(3+((Wxxx-0.31)/1000)^nCW3)/(1+(Hxxx/KCH1)^nCH1)')

sam.add_syn2model('w', 'L26', '(' + 'k0WL/(1+(Lxxx/KWL)^nWL)' +'*'+ 'k0WC/(1+(Cxxx/KWC)^nWC)' + ') *' + 'k0WW*(aw0+(Wxxx/KWW)^nWW/(1+(Wxxx/KWW)^nWW))')
#sam.add_syn2model('w', 'L26', '('  + 'k0WC/(1+(Cxxx/KWC)^nWC)' + ') *' + 'k0WW*(aw0+(Wxxx/KWW)^nWW/(1+(Wxxx/KWW)^nWW))')
sam.add_syn2model('H', 'max_HAM', 'kH')#this is the Ham
#sam.add_syn2model('w', 'L26', '(' + 'k0WL/(1+(Lxxx/KWL)^nWL)' +'+'+ 'k0WC/(1+(Cxxx/KWC)^nWC)' + ') * 1.2')

sam.add_diff2model('L')
sam.add_diff2model('C')
sam.add_diff2model('W')
sam.add_diff2model('H')

# If needed, check equations: ###for setting cutoff for neighber pts, some neighber pts get too far, eg, 0 and 18###!!!!
print(sam.model_str)
```
